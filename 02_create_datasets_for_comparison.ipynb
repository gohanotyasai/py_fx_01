````python
import pandas as pd
import numpy as np
import csv
import os
````

# FXのデータセット（1分間隔）を基に5分間隔、10分間隔にまとめたデータセットを作りCSVファイルとして出力する / 
## 定数を定義

````python
hdCurr = 'curr'
hdOpen = 'curr_OPEN'
hdClose = 'curr_CLOSE'
hdHigh = "curr_HIGH"
hdLow = "curr_LOW"
hdDifOC = 'curr_dif_OC'
hdDifHL = 'curr_dif_HL'
hdIndex = 'DateTime'
hdPip = 'pips_'
hdsOrg = [hdIndex, hdOpen, hdHigh, hdLow, hdClose,'Volume']
hdsCur = [hdIndex, hdOpen, hdClose, hdHigh, hdLow]
hdsDif = [hdIndex, hdDifOC, hdDifHL]

folder_in = './data_in/'
folder_out = './data_out/'
currencyPair =['usdjpy', 'gbpjpy', 'gbpusd']
````

## FXデータセットをdataframeに読み込み、高値安値の差、始値終値の差を計算し、CSVファイルを生成 / 

````python
for cp in currencyPair:
    dfTmp = pd.read_csv(folder_in + cp + ".csv", header=None, names=hdsOrg, sep=';')
    dfTmp[hdDifOC] = dfTmp[hdClose] - dfTmp[hdOpen] 
    dfTmp[hdDifHL] = dfTmp[hdHigh] - dfTmp[hdLow] 
    dfTmp.to_csv(folder_out + cp + ".csv", index=False)
````

## csvファイルを値動き情報と、差情報とに分けdataframeに読み込み / 

````python
dfCur = pd.DataFrame()
dfDif = pd.DataFrame()

for cp in currencyPair:
    dfTmp = pd.read_csv(folder_out + cp + ".csv", sep=',')

    dfTmp_cur = dfTmp[hdsCur].copy()
    dfTmp_cur.columns = [s.replace(hdCurr, cp) for s in dfTmp_cur.columns]
    dfTmp_cur[hdIndex] = pd.to_datetime(dfTmp_cur[hdIndex])
    dfTmp_cur.set_index(hdIndex, inplace=True)
    dfCur = pd.concat([dfCur, dfTmp_cur], axis=1, join='outer')    
    
    dfTmp_dif = dfTmp[hdsDif].copy()
    dfTmp_dif.columns = [s.replace(hdCurr, cp) for s in dfTmp_dif.columns]
    dfTmp_dif[hdIndex] = pd.to_datetime(dfTmp_dif[hdIndex])
    dfTmp_dif.set_index(hdIndex, inplace=True)    
    dfDif = pd.concat([dfDif, dfTmp_dif], axis=1, join='outer')    
````

````python
display(dfCur)
display(dfDif)
````

## オリジナルのデータセットは1分間隔になっている。5分間隔と10分間隔も作成する。なお、値動きに関しては平均値を、差分に関しては合計値を集計する。 / 

````python
dfCur05M = dfCur.groupby(pd.Grouper(freq='5Min')).mean()
dfDif05M = dfDif.groupby(pd.Grouper(freq='5Min')).sum()
dfCur10M = dfCur.groupby(pd.Grouper(freq='10Min')).mean()
dfDif10M = dfDif.groupby(pd.Grouper(freq='10Min')).sum()
````

````python
display(dfCur05M)
display(dfDif05M)
display(dfCur10M)
display(dfDif10M)
````

## 連番を振る（不要かも） / 

````python
dfCur['No'] = range(1, len(dfCur.index) + 1)
dfDif['No'] = range(1, len(dfDif.index) + 1)
dfCur05M['No'] = range(1, len(dfCur05M.index) + 1)
dfDif05M['No'] = range(1, len(dfDif05M.index) + 1)
dfCur10M['No'] = range(1, len(dfCur10M.index) + 1)
dfDif10M['No'] = range(1, len(dfDif10M.index) + 1)
````

## CSVファイル出力 / 

````python
if not os.path.exists(folder_out):
    os.mkdir(folder_out)
    
dfCur.to_csv(folder_out + '01_Curr01M.csv', index=True)
dfDif.to_csv(folder_out + '01_Diff01M.csv', index=True)

dfCur05M.to_csv(folder_out + '05_Curr05M.csv', index=True)
dfDif05M.to_csv(folder_out + '05_Diff05M.csv', index=True)

dfCur10M.to_csv(folder_out + '10_Curr10M.csv', index=True)
dfDif10M.to_csv(folder_out + '10_Diff10M.csv', index=True)
````

# 比較しやすくするため、データを加工する / 

## 【標準化 / Standardization】SciPyを用いてz値（平均0、分散1に正規化）を求める。個々のデータが平均値から標準偏差いくつぶん離れているかを計算。

````python
import scipy.stats
def adjustVal(df):
    dfTmp = df.copy()
    for hd in dfTmp.columns:
        if hd != "No":
            dfTmp[hd] = scipy.stats.zscore(dfTmp[hd], nan_policy='omit')
    return dfTmp
````

### scikit-learnを用いた正規化（Min-Max Normalization）も試したが、標準化した結果を比較した方が見やすかった。/
```
from sklearn import preprocessing
def adjustVal(df):
    dfTmp = df.copy()
    for hd in dfTmp.columns:
        if hd != "No":
            dfTmp[hd] = preprocessing.minmax_scale(dfTmp[hd])
    return dfTmp
```

## データセットを標準化 / 

````python
dfCur_adj = adjustVal(dfCur)
dfDif_adj = adjustVal(dfDif)
dfCur05M_adj = adjustVal(dfCur05M)
dfDif05M_adj = adjustVal(dfDif05M)
dfCur10M_adj = adjustVal(dfCur10M)
dfDif10M_adj = adjustVal(dfDif10M)
````

````python
display(dfCur_adj)
display(dfDif_adj)
````

## CSVファイル出力 / 

````python
dfCur_adj.to_csv(folder_out + '01_Curr01M_adj.csv', index=True)
dfDif_adj.to_csv(folder_out + '01_Diff01M_adj.csv', index=True)

dfCur05M_adj.to_csv(folder_out + '05_Curr05M_adj.csv', index=True)
dfDif05M_adj.to_csv(folder_out + '05_Diff05M_adj.csv', index=True)

dfCur10M_adj.to_csv(folder_out + '10_Curr10M_adj.csv', index=True)
dfDif10M_adj.to_csv(folder_out + '10_Diff10M_adj.csv', index=True)
````
