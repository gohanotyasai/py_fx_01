```python
from histdata import download_hist_data as dl
from histdata.api import Platform as P, TimeFrame as TF
import shutil
import pandas as pd
import numpy as np
import csv
import os
```

```python
hdOpen = 'curr_OPEN'
hdClose = 'curr_CLOSE'
hdCurr = 'curr'
hdIndex = 'DateTime'
hdPip = 'pips_'
hd = [hdIndex,hdOpen,'HIGH','LOW',hdClose,'Volume']
hd2 = [hdIndex, hdOpen, hdClose]
```

```python
folder = './data/'

y = '2022'
mon = '5'
currencyPair =['usdjpy', 'gbpjpy', 'gbpusd']

dfVal = pd.DataFrame()
dfInc = pd.DataFrame()

for cp in currencyPair:
    zipfile = dl(year=y, month=mon, pair=cp, platform=P.GENERIC_ASCII, time_frame=TF.ONE_MINUTE, output_directory=folder)
    csvfile = zipfile.replace('zip', 'csv')
    shutil.unpack_archive(zipfile, folder)
    dfTmp = pd.read_csv(csvfile, header=None, names=hd, sep=';')

    dfTmp = dfTmp[hd2]
    dfTmp.columns = [s.replace(hdCurr, cp) for s in dfTmp.columns]
    
    dfTmp[hdIndex] = pd.to_datetime(dfTmp[hdIndex])
    dfTmp.set_index(hdIndex, inplace=True)
    dfTmp = dfTmp.groupby(pd.Grouper(freq='10Min')).mean()
    
    dfVal = pd.concat([dfVal, dfTmp], axis=1, join='outer')

    dfTmp[hdPip + cp] = dfTmp[hdClose.replace(hdCurr, cp)] - dfTmp[hdOpen.replace(hdCurr, cp)]
    dfTmp = dfTmp.groupby(pd.Grouper(freq='10Min')).sum()
    
    dfInc = pd.concat([dfInc, dfTmp[hdPip + cp]], axis=1, join='outer')
```
