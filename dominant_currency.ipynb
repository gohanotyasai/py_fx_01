```python
from histdata import download_hist_data as dl
from histdata.api import Platform as P, TimeFrame as TF
import shutil
import pandas as pd
import numpy as np
import csv
import os
```

```python
hdCurr = 'curr'
hdOpen = 'curr_OPEN'
hdClose = 'curr_CLOSE'
hdHigh = "curr_HIGH"
hdLow = "curr_LOW"
hdDifOC = 'curr_dif_OC'
hdDifHL = 'curr_dif_HL'
hdIndex = 'DateTime'
hdPip = 'pips_'
hdsOrg = [hdIndex, hdOpen, hdHigh, hdLow, hdClose,'Volume']
hdsCur = [hdIndex, hdOpen, hdClose, hdHigh, hdLow]
hdsDif = [hdIndex, hdDifOC, hdDifHL]

folder = './data/'
folder_out = './data_out/'
```

```python
y = '2022'
mon = '5'
currencyPair =['usdjpy', 'gbpjpy', 'gbpusd']

for cp in currencyPair:
    os.remove(folder + cp + ".csv")

for cp in currencyPair:
    zipfile = dl(year=y, month=mon, pair=cp, platform=P.GENERIC_ASCII, time_frame=TF.ONE_MINUTE, output_directory=folder)
    shutil.unpack_archive(zipfile, folder)
    csvfile = zipfile.replace('zip', 'csv')
    os.rename(csvfile, folder + cp + ".csv")
```

```python
for cp in currencyPair:
    dfTmp = pd.read_csv(folder + cp + ".csv", header=None, names=hdsOrg, sep=';')
    dfTmp[hdDifOC] = dfTmp[hdClose] - dfTmp[hdOpen] 
    dfTmp[hdDifHL] = dfTmp[hdHigh] - dfTmp[hdLow] 
    dfTmp.to_csv(folder_out + cp + ".csv", index=False)
```

```python
dfCur = pd.DataFrame()
dfDif = pd.DataFrame()

for cp in currencyPair:
    dfTmp = pd.read_csv(folder_out + cp + ".csv", sep=',')

    dfTmp_cur = dfTmp[hdsCur].copy()
    dfTmp_cur.columns = [s.replace(hdCurr, cp) for s in dfTmp_cur.columns]
    dfTmp_cur[hdIndex] = pd.to_datetime(dfTmp_cur[hdIndex])
    dfTmp_cur.set_index(hdIndex, inplace=True)
    dfCur = pd.concat([dfCur, dfTmp_cur], axis=1, join='outer')    
    
    dfTmp_dif = dfTmp[hdsDif].copy()
    dfTmp_dif.columns = [s.replace(hdCurr, cp) for s in dfTmp_dif.columns]
    dfTmp_dif[hdIndex] = pd.to_datetime(dfTmp_dif[hdIndex])
    dfTmp_dif.set_index(hdIndex, inplace=True)    
    dfDif = pd.concat([dfDif, dfTmp_dif], axis=1, join='outer')    
```

```python
display(dfCur)
display(dfDif)
```

```python
dfCur10M = dfCur.groupby(pd.Grouper(freq='10Min')).mean()
dfDif10M = dfDif.groupby(pd.Grouper(freq='10Min')).sum()
```

```python
display(dfCur10M)
display(dfDif10M)
```

```python
dfCur['No'] = range(1, len(dfCur.index) + 1)
dfDif['No'] = range(1, len(dfDif.index) + 1)
dfCur10M['No'] = range(1, len(dfCur10M.index) + 1)
dfDif10M['No'] = range(1, len(dfDif10M.index) + 1)
```

```python
if not os.path.exists(folder_out):
    os.mkdir(folder_out)
    
dfCur.to_csv(folder_out + '11_Curr.csv', index=True)
dfCur10M.to_csv(folder_out + '12_Curr10M.csv', index=True)

dfDif.to_csv(folder_out + '21_Diff.csv', index=True)
dfDif10M.to_csv(folder_out + '22_Diff10M.csv', index=True)
```

```python
def adjustVal(df, tgtHd, ratio):
    dfTmp = df.copy()
    for hd in dfTmp.columns:
        if tgtHd in hd:
            dfTmp[hd] = dfTmp[hd] * ratio
    return dfTmp
```

```python
dfCur_adj = adjustVal(dfCur, 'gbpusd', 100)
dfDif_adj = adjustVal(dfDif, 'gbpusd', 100)
dfCur10M_adj = adjustVal(dfCur10M, 'gbpusd', 100)
dfDif10M_adj = adjustVal(dfDif10M, 'gbpusd', 100)
```

```python
dfCur_adj.to_csv(folder_out + '11_Curr_adj.csv', index=True)
dfCur10M_adj.to_csv(folder_out + '12_Curr10M_adj.csv', index=True)
dfDif_adj.to_csv(folder_out + '21_Diff_adj.csv', index=True)
dfDif10M_adj.to_csv(folder_out + '22_Diff10M_adj.csv', index=True)
```
